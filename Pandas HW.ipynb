```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

school_data_complete
```

```python
#Use prior group to now get average math score by school
grouped_top_perform_df = top_perform_df.groupby(["school_name"])
ave_math_score = grouped_top_perform_df["math_score"].mean()
ave_math_score
```

```python
#Use prior group to now get average reading score by school
grouped_top_perform_df = top_perform_df.groupby(["school_name"])
ave_reading_score = grouped_top_perform_df["reading_score"].mean()
ave_reading_score
```

```python
#Get total number of math scores by school 
grouped_top_perform_df = top_perform_df.groupby(["school_name"])
total_math_scores = grouped_top_perform_df["math_score"].count()
total_math_scores.head(5)
```

```python
#Get total number of math scores greater than or equal to 70

passing_math_scores = top_perform_df[top_perform_df['math_score'] >= 70].groupby('school_name')['math_score'].count()
passing_math_scores.head(5)
```

```python
#Divide last two cells to get % of students that passed math test
perc_passing_math = passing_math_scores/total_math_scores*100
perc_passing_math.head(5)
```

```python
#Get total number of reading scores by school 
grouped_top_perform_df = top_perform_df.groupby(["school_name"])
total_reading_scores = grouped_top_perform_df["reading_score"].count()
total_math_scores.head(5)
```

```python
#Get total number of reading scores greater than or equal to 70
passing_reading_scores = top_perform_df[top_perform_df['reading_score'] >= 70].groupby('school_name')['reading_score'].count()
passing_reading_scores.head(5)
```

```python
#Divide last two cells to get % of students that passed reading test
perc_passing_reading = passing_reading_scores/total_reading_scores*100
perc_passing_reading.head(5)
```

```python
#Overall Passing Rate (Average of the above two)
overall_passing_rate = (perc_passing_reading+perc_passing_math)/2
overall_passing_rate.head()
```

```python
#Build resulting dataframe table
summary_performing_data = pd.DataFrame({"Total Schools":total_schools, "Total Students":total_students,
                            "Total Budget":total_budget, "Average Math Score":[ave_math_score],
                            "Average Reading Score":[ave_reading_score],"% Passing Math":math_passing_perc, 
                            "Overall Passing Rate":[overall_passing_rate]})
summary_performing_data
```
